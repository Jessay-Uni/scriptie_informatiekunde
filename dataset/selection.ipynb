```python
# Author: Jessay Beukema
import pandas
```

```python
# Translates .csv to pandas dataframe
articles_df = pandas.read_csv("dutch-news-articles.csv")

articles_df
```

```python
# Samples 1000 random articles from dataframe. Uses seed for reproducability
sample_articles_df = articles_df.sample(n=1000, axis=0, random_state=42)

sample_articles_df
```

```python
# Stores sample as .csv
# This file was the corpus used for doing annotations
sample_articles_df.to_csv("dataset.csv")
```

```python
# Removes sampled articles from full database
articles_2_df = pandas.concat([articles_df, sample_articles_df, sample_articles_df]).drop_duplicates(keep=False)
```

```python
# Samples 500 random articles, using a seed for reproducability
# This second sample doesn't contain any article that was present in the first sample
sample_articles_2_df = articles_2_df.sample(n=500, axis=0, random_state=420)

sample_articles_2_df
```

```python
# Stores corpus used for the experiment
sample_articles_2_df.to_csv("data_experiment.csv")
```

```python
experiment_1 = sample_articles_2_df.sample(n=250, axis=0, random_state=7)

experiment_1
```

```python
experiment_2 = pandas.concat([sample_articles_2_df, experiment_1, experiment_1]).drop_duplicates(keep=False)

experiment_2
```

```python
experiment_1.to_csv("experiment-1.csv")
experiment_2.to_csv("experiment-2.csv")
```
